```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('Online Sales Data.csv')

# Display the first few rows of the dataset
print(data.head())
```

```python
data.head()
```

```python
data.info()
```

```python
# Calculate total revenue
total_revenue = data['Total Revenue'].sum()
print(f"Total Revenue: ${total_revenue:.2f}")
```

```python
# Find the best-selling product by units sold
best_selling_product = data.loc[data['Units Sold'].idxmax()]
print(f"Best Selling Product: {best_selling_product['Product Name']} with {best_selling_product['Units Sold']} units sold")
```

```python
# Analyze sales by region
sales_by_region = data.groupby('Region')['Total Revenue'].sum()
print("Sales by Region:")
print(sales_by_region)
```

```python
# Analyze sales by product category
sales_by_category = data.groupby('Product Category')['Total Revenue'].sum()
print("Sales by Product Category:")
print(sales_by_category)
```

```python
# Analyze sales by payment method
sales_by_payment = data.groupby('Payment Method')['Total Revenue'].sum()
print("Sales by Payment Method:")
print(sales_by_payment)
```

```python
# Find the most expensive product sold
most_expensive_product = data.loc[data['Unit Price'].idxmax()]
print(f"Most Expensive Product Sold: {most_expensive_product['Product Name']} at ${most_expensive_product['Unit Price']:.2f}")
```

```python
# Find the average unit price by product category
avg_price_by_category = data.groupby('Product Category')['Unit Price'].mean()
print("Average Unit Price by Product Category:")
print(avg_price_by_category)
```

```python
# Find the total units sold by product category
units_sold_by_category = data.groupby('Product Category')['Units Sold'].sum()
print("Total Units Sold by Product Category:")
print(units_sold_by_category)
```

```python
# Find the top 5 products by revenue
top_5_products_by_revenue = data.groupby('Product Name')['Total Revenue'].sum().nlargest(5)
print("Top 5 Products by Revenue:")
print(top_5_products_by_revenue)
```

```python
# Find the top 5 regions by revenue
top_5_regions_by_revenue = data.groupby('Region')['Total Revenue'].sum().nlargest(5)
print("Top 5 Regions by Revenue:")
print(top_5_regions_by_revenue)
```

```python
# Find the top 5 payment methods by revenue
top_5_payment_methods_by_revenue = data.groupby('Payment Method')['Total Revenue'].sum().nlargest(5)
print("Top 5 Payment Methods by Revenue:")
print(top_5_payment_methods_by_revenue)
```
